```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import pytraj as pt

traj = pt.load('Sim/md.crd', top='Sim/ArM_solv.prmtop')
```

```python
data_rmsd_first = pt.rmsd(traj, ref=0, mask=":1-13&!@H*")
print(data_rmsd_first)
```

```python
%matplotlib inline
```

```python
plt.plot(data_rmsd_first)
plt.xlabel('Frame')
plt.ylabel('RMSD(Angstrom)')
plt.suptitle("RMSD of Protein")
plt.savefig('RMSD.png', bbox_inches='tight', dpi=300)
```

```python
mat = pt.pairwise_rmsd(traj, mask=":1-13&!@H*", frame_indices=range(0, 500, 10))
```

```python
plt.imshow(mat, cmap='viridis')
plt.xlabel('Frame')
plt.ylabel('Frame')
plt.colorbar(label=r'RMSD ($\AA$)')
plt.suptitle("RMSD of Protein")
plt.savefig('2D_RMSD.png', bbox_inches='tight', dpi=300)
```
